### Lamine Yamal Passmap

```python
# players who played in the final
df['player'].unique()
```

```python
Lamine_Yamal = df[df['player'] == 'Lamine Yamal Nasraoui Ebana']
Lamine_Yamal
```

```python
Lamine_Yamal['type'].value_counts()
```

```python
Lamine_Yamal = Lamine_Yamal[Lamine_Yamal['type'] == 'Pass']
Lamine_Yamal
```

```python
Lamine_Yamal = Lamine_Yamal[['location', 'pass_end_location', 'pass_outcome']]
Lamine_Yamal
```

```python
Lamine_Yamal[['x','y']] = Lamine_Yamal['location'].apply(pd.Series)
Lamine_Yamal[['end_x','end_y']] = Lamine_Yamal['pass_end_location'].apply(pd.Series)
Lamine_Yamal['pass_outcome'] = Lamine_Yamal['pass_outcome'].fillna('Complete')
Lamine_Yamal
```

```python
Lamine_Yamal['pass_outcome'].value_counts()
```

```python
Lamine_Yamal['final_third_passes'] = (Lamine_Yamal['end_x'] >= 80) & (Lamine_Yamal['x'] < 80)
```

```python
Lamine_Yamal['final_third_passes'].value_counts()
```

```python
#Lamine_Yamal = Lamine_Yamal[Lamine_Yamal['final_third_passes'] == True]
```

```python
Lamine_Yamal
```

```python
from mplsoccer import Pitch

pitch = Pitch(pitch_type = 'statsbomb', pitch_color='grass')
fig, ax = pitch.draw()

for x in Lamine_Yamal.to_dict(orient='records'):
    if x['final_third_passes'] == True and x['pass_outcome'] == 'Complete':
        pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=5, transparent=True, comet=True, ax=ax, color='purple')
    elif x['final_third_passes'] == True and x['pass_outcome'] != 'Complete':
        pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=5, transparent=True, comet=True, ax=ax, color='b')
    elif  x['pass_outcome'] == 'Complete':
        pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=5, transparent=True, comet=True, ax=ax, color='w')
    elif x['pass_outcome'] == 'Incomplete':
        pitch.lines(x['x'], x['y'], x['end_x'], x['end_y'], lw=5, transparent=True, comet=True, ax=ax, color='r')
        
ax.set_title("Lamine Yamal Passes in the Euro 2024 Final", fontsize=18, fontfamily='monospace')

from matplotlib.lines import Line2D

legend_elements = [
    Line2D([0], [0], color='w', lw=4, label='Completed Pass'),
    Line2D([0], [0], color='r', lw=4, label='Incompleted Pass'),
    Line2D([0], [0], color='purple', lw=4, label='Completed Final Third Pass'),
    Line2D([0], [0], color='b', lw=4, label='Incompleted Final Third Pass')
]

ax.legend(handles=legend_elements, loc='best')

fig.savefig("D:\\Euro 2024 Final - Lamine Yamal Passmap.png", dpi=300)
```

### Bar Chart for Total Passes

```python
df['player'].unique()
```

```python
# Filtering to get only the passes of the players
Passes = df[df['type'] == 'Pass']
Passes
```

```python
#Specifying the columns required
Passes = Passes[['player','type','pass_outcome']]
Passes
```

```python
#Filling the nan values as 'Complete'
Passes['pass_outcome'] = Passes['pass_outcome'].fillna('Complete')
Passes
```

```python
Passes['pass_outcome'].value_counts()
```

```python
# making 2 new columns for incompleted and completed passes
Passes['completed_passes'] = (Passes['pass_outcome'] == 'Complete')
Passes['incompleted_passes'] = (Passes['pass_outcome'] != 'Complete')
Passes
```

```python
# quantifying the data
Passes['completed_passes'] = Passes['completed_passes'].replace({True:1, False:0})
Passes['incompleted_passes'] = Passes['incompleted_passes'].replace({True:1, False:0})
Passes
```

```python
# replacing the long names with shorter names
Passes['player'] = Passes['player'].replace({"Daniel Carvajal Ramos":"Dani Carvajal","Daniel Olmo Carvajal":"Dani Olmo",
                                        "Fabián Ruiz Peña":"Fabián Ruiz","José Ignacio Fernández Iglesias":"Nacho Fernandez",
                                        "Lamine Yamal Nasraoui Ebana":"Lamine Yamal","Marc Cucurella Saseta":"Marc Cucurella",
                                        "Martín Zubimendi Ibáñez":"Martín Zubimendi","Mikel Oyarzabal Ugarte":"Mikel Oyarzabal",
                                        "Nicholas Williams Arthuer":"Nicholas Williams","Robin Aime Robert Le Normand":"Robin Le Normand",
                                        "Rodrigo Hernández Cascante":"Rodri","Unai Simón Mendibil":"Unai Simón",
                                        "Álvaro Borja Morata Martín":"Álvaro Morata"})
Passes['player']
```

```python
Passes.dtypes
```

```python
# creating a dataframe for incompleted and completed passes with the player names
Passes = Passes[['player','completed_passes','incompleted_passes']]
Passes

#unable to see the whole dataset, saving the data in a csv format to view the entire datset and deciding a match to analyse
#Passes.to_csv('Passes.csv', index=False)
```

```python
Passes.value_counts()
```

```python
# grouping the player names and summing the passes that they have attempted
Passes = Passes.groupby(['player'])[['completed_passes','incompleted_passes']].sum().reset_index()
Passes
```

#### horizontal bar graph

```python
#visualizing the passes attempted by the players using horizontal bar graph

def add_labels_horizontal(x, y1, y2):
    for i in range(len(x)):
        plt.text(y1[i] / 2, i, int(y1[i]), ha='center', va='center', color='w', fontweight='bold')
        plt.text(y1[i] + y2[i] / 2, i, int(y2[i]), ha='center', va='center', color='k')

plt.figure(figsize=(10, 8))

x = Passes['player']
y1 = Passes['completed_passes']
y2 = Passes['incompleted_passes']

plt.barh(x, y1, color='limegreen', label="Completed Passes")
plt.barh(x, y2, left=Passes['completed_passes'], color='lightcoral', label="Incompleted Passes")

add_labels_horizontal(x, y1, y2)

plt.xlabel("Number of Passes Attempted")
plt.ylabel("Player")
plt.title("Pass Attempts by Players - Euro 2024 Final")
plt.legend()
plt.tight_layout()

plt.savefig("D:\\Euro 2024 Final Passes.png", dpi=300)

plt.show()
```

#### vertical bar graph

```python
#visualizing the passes attempted by the players using vertical bar graph
def add_labels(x, y1, y2):
    for i in range(len(x)):
        plt.text(i, y1[i] // 2, y1[i], ha='center', va='center', color='k')
        plt.text(i, y1[i], y2[i], ha='center')

plt.figure(figsize=(10, 6))

x = Passes['player']
y1 = Passes['completed_passes']
y2 = Passes['incompleted_passes']

plt.bar(x, y1, color='limegreen', label="completed_passes")
plt.bar(x, y2, bottom=y1, color='lightcoral', label="incompleted_passes")
plt.xticks(rotation=90)

add_labels(x, y1, y2)

plt.title("passes attempted by the players in the final")
plt.ylabel("no. of passes attempted")
plt.xlabel("player")
plt.legend()

plt.show()
```

### Shotmap

```python
Shotmap = df[df['type'] == 'Shot']
Shotmap
```

```python
Shotmap = Shotmap[['team','player','location','shot_outcome','shot_statsbomb_xg']].reset_index(drop=True, inplace=False)
Shotmap
```

```python
Shotmap[['x','y']] = Shotmap['location'].apply(pd.Series)
Shotmap
```

```python
# Let's create our shotmap
from mplsoccer import Pitch

pitch = Pitch(pitch_type='statsbomb', pitch_color='#aabb97', line_color='white',
              stripe_color='#c2d59d', stripe=True)
fig, ax = pitch.draw()

# plot the shots
# We want the size of the shot to be 500 * the expected goals value
# We also want to leave the marker empty if the shot was missed and fill it if it was a goal

for x in Shotmap.to_dict(orient='records'):
    if x['team'] == 'Spain':
        # We want to plot France on the left side of the pitch
        # So we need to mirror both the x and y coordinates
        pitch.scatter(x=120-x['x'], y=80-x['y'], ax=ax, s=500*x['shot_statsbomb_xg'], ec='black', c='blue' if x['shot_outcome'] == 'Goal' else 'white')
    else:
        pitch.scatter(x=x['x'], y=x['y'], ax=ax, s=500*x['shot_statsbomb_xg'], ec='black', c='blue' if x['shot_outcome'] == 'Goal' else 'white')
        

# We can add a title above the different halves of the pitch
ax.text(30, 86, 'Spain', ha='center', fontsize=20)
ax.text(90, 86, 'England', ha='center', fontsize=20)

# We can also add a title directly above the text we added
ax.set_title('Euro 2024 Final Shotmap', fontsize=20)

# Lets also add a legend for the size of the shots in the bottom middle of the pitch
pitch.scatter(55, 77.5, s=500*.05, ec='black', c='white', ax=ax)
pitch.scatter(59.5, 77.5, s=500*.15, ec='black', c='white', ax=ax)
pitch.scatter(65, 77.5, s=500*.3, ec='black', c='white', ax=ax)

# We can add a label for the legend
ax.text(60, 74, 'Size represents xG', ha='center', fontsize=8, fontfamily='monospace')

fig.savefig("D:\\Euro 2024 Final Shotmap.png", dpi=300)
```
